```python
import os
```

```python
#list all the folders and files in the directory
curr_dir = os.getcwd()
print("Current Directory:", curr_dir)
```

```python
for root,dirs,files in os.walk(curr_dir):
    print("Root:", root)
    print("Directories:", dirs)
    for dir in dirs:
        print("Directory:", dir)
    print("Files:", files)
    for file in files:
        print("File:", file)
    print("\n")
```
